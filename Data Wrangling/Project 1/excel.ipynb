## 3. Parse Excel File


__First step is to parse the excel file.__

Open Excel file by creating a Pandas ExcelFile object, dataEx.

```python
dataEx = pd.ExcelFile('basic_indicators.xlsx')
dataEx
```

In some cases, particularly while an Excel file has a lot of worksheet, it may be good to view all the sheets by their sheet names. `sheet_names` is used to view all the sheet names. And then, parse the specific sheet by using `df.parse('sheet name')`.

```python
# check sheet names in the excel file
# only one sheet
dataEx.sheet_names
```

```python
# read 'Basic Indicators' sheet into pandas dataframe
df = dataEx.parse('Basic Indicators')
df.head(10)
```

We can see that function `pandas.ExcelFile()` can only read the file into main memory and it cannot parse the work sheet directly. In this case, `ExcelFile` can only read the excel file into main memory location, _0x1e01b476898_, even there is only one work sheet, _Basic Indicators_. Next step is to parse the particular work sheet using sheet name. The result will be dataframe which we need.

==========================================================================

__The next step is to wrangle the data.__

From dataframe before, the first four row is useless and it can be droped using function `dataframe.drop()`.

```python
# drop first four useless rows
df = df.drop(df.index[:4])
df.head()
```

Check all the columns in this dataframe. Make sure the unvisualable space in each column label.

```python
df.columns
```

From the excel data, there are some rows showing the summary information of this dataset. 
However, we only need detail information before the summary. Thus we should get the index 
of summary and then drop all the information below summary. `index.item()` is used to get the number of the index.

```python
# Check the index of 'SUMMARY' item in column 'TABLE 1. BASIC INDICATORS'
summaryIndex = df[df['TABLE 1. BASIC INDICATORS'] == 'SUMMARY'].index.item()
print(summaryIndex)
# Check the index of last row in dataframe
lastIndex = df.tail(1).index.item()
print(lastIndex)
```

The indexes of "SUMMARY" and last row were got. These rows are useless for this dataframe. All the row between these two should be dropped.

```python
# Drop the rows from SUMMARY to last
df.drop(list(range(summaryIndex, lastIndex+1)), inplace = True)
df
```

For the final dataframe, there is no column to record 'x'. Thus, we should drop the column of 'x'. First, the charactor 'x' can be replced by NaN use regular expression.

Next, we can used function `dropna()` to drop all the NA rows and columns.  `how = 'all'` means drop the columns or rows if there are all NA.

And then, replace the '-' cell with nothing using regular expression.

```python
# replace single 'x' with NaN using regular expression
df.replace(r'^x$', np.nan, inplace = True, regex = True)
df
```

```python
# drop all NAN rows and columns
df = df.dropna(1, how = 'all').dropna(0, how = 'all')
df
```

```python
# replace '-' with empty using re
df = df.replace(r'^–$', '', regex = True)
df
```

From the dataframe before, the column name is not what we want. And for better readable, the first column, country, can be set as index and replace the index name as Country Name. Then, it will be easy to rename the other columns with the number from 0.

```python
# Set the first colums as index 
df = df.set_index(df['TABLE 1. BASIC INDICATORS'].values)
df.index.name = 'Country Name'
```

```python
# drop the duplicate column which is not index column 
df.drop('TABLE 1. BASIC INDICATORS', 1, inplace = True)
df
```

```python
df.columns = list(range(len(df.columns)))
df.head()
```

By checking the data type of the dataframe, we can find out that all the values are recorded as object. Thus, we should convert the object type into number. `dataframe.convert_objects(convert_numeric = True)` is used to convert the object type to numeric type if the column can be convert.

```python
# Check the data type of each column in the dataframe
df.dtypes
```

```python
# convert object data type into numeric type
df = df.convert_objects(convert_numeric = True)
df.dtypes
```

```python
# export dataframe into csv file with no decimal float format
df.to_csv('./basic_indicators.csv', float_format='%.0f', encoding = 'utf-16', sep = '\t')
```
