We can found that, all the numbers are in the same range of 0 to 100. This feature could be used in regular expression. Moreover, a item can also be '-'. Thus, I define the regular expression pattern as `r'–|100|[1-9]?\dx?'` to seperate the merged items except country name. Because there are a lot of part which merged together. I have defined a function for seperation.

````python
# define the function to seperate the merged numbers
def divideMerge(ls, index):
    """
    Divide the merged values. All the merged item are numbers between 0 to 100.
    
    Arguments:
    ls -- a list of data which records a line of one country information.
    index -- the index of the input item in ls
    
    Return:
    ls -- the renew ls
    """
    matchLt = re.findall(r'–|100|[1-9]?\dx?', ls[index]) # use regular expression to 
                                                          # find all data match the pattern
    if len(matchLt) > 1:  # there are more than one items merged together
        ls[index : index+1] = matchLt  # renew the ls by replace the checked item with the new list of items
        print(matchLt)
    return ls
````

````python
# iterate the newData list use for loop
# 1. seperate the merged country
# 2. seperate the merged numbers
# 3. neglect 'x' followed by number
for i in newData:
    # check if the country item are merged with ather values. 
    # If yes, find out the merged items. If not, continue.
    matchStr = re.search(r'^([a-zA-Z\s(\)]+)(100|[0-9]{2}|–)(100|[0-9]{2}|–)(100|[0-9]{2}|–)$', i[0])
    if matchStr != None:  # country is merged.
        i[0:1] = matchStr.group(1), matchStr.group(2), matchStr.group(3), matchStr.group(4)
    
    # check items without country.
    while True:
        if len(i) == 23:  # if there are 23 items in i, no merged item.
            break
        for j in range(1, len(i)): # check each item in i and find out if the item merged.
            i = divideMerge(i, j)  # used the function defined above.
        print(i)
        print('New length :', len(i))
    
    # neglect 'x' followed by number
    matchX = re.findall(r'\d+x', str(i)) # find all the item contains 'x'
    if len(matchX) != 0:
        for j in matchX:
            idx = i.index(j)
            i[idx] = i[idx][:-1] # neglect 'x'
        
````

### Check the wrangled result
There are some weak point by using the regular expression pattern `r'–|100|[1-9]?\dx?'`. This pattern can seperate the merged string into several number between 0 to 100. However, it cannot recognize whether the merged value is 'DD' and 'D' or 'D' and 'DD'. If the value is 'DD' and 'D' the pattern will certainly correct. But if the value is 'D' and 'DD', it will misseperate into 'DD' and 'D'. In this part, we will focus on this problem.

__Check problems 1 :__

The merged value has been divided through regular expression. This step is to check if the division gives a correct value. The column[4] and column[5] was merged before. We find that the value in column[4] must in the range between column[5] and column[6].

````python
# Check of the abnormal line
def checkTUR(newData, urbanIndex):
    """
    Check of the abnormal line in total, urban, rural:
    if the total value is smaller than min(urban, rural) or 
    larger than max(urban, rural), the line will be seen as abnormal.
    
    Arguments:
    newData -- the wrangled total data
    urbanIndex -- the index of urban data in each line of newData
    
    Return:
    errorList -- the dictionary {index : item} of obnormal line 
    """
    errorList = dict()
    for index, i in enumerate(newData):
        if '–' not in [i[urbanIndex-1], i[urbanIndex], i[urbanIndex+1]]:
            urban = int(i[urbanIndex])
            rural = int(i[urbanIndex+1])
            total = int(i[urbanIndex-1])
            if total < min(urban, rural) or total > max(urban, rural):
                errorList[str(index)] = i
    return errorList
````
