## Modify Dataset2 to adopt schema in Dataset1
### Modify Contract Type
Column 'Contract' Type contains NaN value

```python
ds2['Contract Type'].value_counts()
```

From data above, we find out that dataset2 only contains two kinds of data: ft, pt. These two kinds of data have the same means in dataset1: full-time and part-time. Thus modify the two names.

```python
ds2['Contract Type'].replace({'ft':'full-time', 'pt':'part-time', 'n/a':'non-specified'}, inplace=True)
ds2.head()
```

```python
# check result
ds2['Contract Type'].value_counts()
```

### Modify Contract Time
Contains NaN values

```python
ds2['Contract Time'].value_counts()
```

```python
ds2['Contract Time'].replace({'perm.':'permanent', 'contr.':'contract', 'n/a':'non-specified'}, inplace=True)
ds2.head()
```

```python
# check result
ds2['Contract Time'].value_counts()
```

### Compare category in two dataset

```python
# dataset1
cate_list1 = ds1.Category.value_counts().index

# dataset 2
cate_list2 = ds2.Category.value_counts().index
```

```python
# check if there is any different categories in these two dataset.
for i in cate_list2:
    if i not in cate_list1:
        print(i)
```

We find out that, the category item 'Finance & Accounting Jobs' are not in dataset1. Check in dataset1, it is wrote as 'Accounting & Finance Jobs'. Thus, we modidy it in dataset2.

```python
ds2.Category.replace({'Finance & Accounting Jobs':'Accounting & Finance Jobs'}, inplace=True)
ds2.Category.value_counts()
```

### Modify the salary column
Comparing dataset1 with dataset2, we find out that the column in dataset1 name as 'Salary per annum'. However, in dataset2, the column is named as 'Salary per month'. We should modify dataset2 into 'Salary per annum'. Thus, we should add a new column for it and drop the 'Salary per month' column.

```python
ds2['Salary per annum'] = ds2['Salary per month'].apply(lambda x: x*12)
ds2.head()
```

```python
# drop salary per month
ds2.drop(['Salary per month'], axis=1, inplace=True)
ds2.head()
```

Convert int64 typy salary into string with string type to make data type consistency.

```python
ds2['Salary per annum'] = ds2['Salary per annum'].apply(lambda x: str(x))
ds2['Salary per annum'].dtype
```

### Modify Date type
Modify data type in dataset 2 to follow the type of dataset 1.

```python
ds2.OpenDate = pd.to_datetime(ds2.OpenDate, errors='coerce')
ds2.query('OpenDate == "NaT"')
```

```python
ds2.CloseDate = pd.to_datetime(ds2.CloseDate, errors='coerce')
ds2.query('CloseDate == "NaT"')
```

```python
ds2.OpenDate = ds2.OpenDate.apply(lambda x: x.strftime('%Y%m%dT%H%M%S'))
ds2.CloseDate = ds2.CloseDate.apply(lambda x: x.strftime('%Y%m%dT%H%M%S'))
ds2.head()
```

## Integrating two datasets
### Remane the column in dataset2

```python
ds1.columns
```

```python
ds2.columns
```

```python
ds2 = ds2.rename(columns = {'Source Name': 'SourceName',
                            'location': 'Location', 
                            'Contract Type': 'ContractType',
                            'Contract Time': 'ContractTime'})
ds2.head(2)
```

### Integrate two datasets

```python
ds_final = pd.merge(ds1,ds2,how='outer',on=ds1.columns.tolist())
```

```python
# for check data structure
print('Final dataset shape:', ds_final.shape)
print('Dataset1 shape:', ds1.shape)
print('Dataset2 shape:', ds2.shape)
```

```python
ds_final.head()
```

```python
# check if id in these two data set are unique
ds_final[ds_final.Id.duplicated(keep=False)]
```

Id column is totally unique in two datasets. Thus, we can set Id as index and it would be easier to judge if other string columns are unique.

```python
ds_final = ds_final.set_index(ds_final['Id'].values)
ds_final.index.name = 'Id'
ds_final.drop('Id', 1, inplace=True)
ds_final.head()
```

```python
ds_final.info()
```
